```python
df['Arrival_Hour'] = pd.to_datetime(df['Arrival_Time'], format='mixed', errors='coerce').dt.hour
df['Arrival_Minute'] = pd.to_datetime(df['Arrival_Time'], format='mixed', errors='coerce').dt.minute
df['Journey_Day']=pd.to_datetime(df['Date_of_Journey'], format='%d-%m-%Y').dt.day
df['Journey_Month']=pd.to_datetime(df['Date_of_Journey'], format='%d-%m-%Y').dt.month
```
