```python
import sys
import matplotlib
matplotlib.use('Agg')

import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

df = pandas.read_csv("data.csv")

d = {'UK': 0, 'USA': 1, 'N': 2}
df['NATIONALITY'] = df['NATIONALITY'].map(d) # Corrected 'Nationality' to 'NATIONALITY'

d = {'YES': 1, 'NO': 0}
df['GO'] = df['GO'].map(d) # Changed 'Go' to 'GO'
print(df)

features = ['Age', 'Experience', 'Rank', 'NATIONALITY'] # Corrected 'Nationality' to 'NATIONALITY'

X = df[features]
y = df['GO'] # Changed 'Go' to 'GO'


dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)

tree.plot_tree(dtree, feature_names=features)
```
